```python
#import thư viện cần thiết
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline 
sns.set(color_codes = True)

#load du lieu tu file
df = pd.read_csv("data.csv")
df.head(5)
df.tail(5)
```

```python
#Kiem tra du lieu thuoc tinh
df.dtypes
```

```python
#xem thong ke cua bo du lieu
df.describe()
```

```python
#Loai bo thuoc tinh du thua
df = df.drop(['Engine Fuel Type'], axis = 1)
df.head(5)
```

```python
#Doi ten thuoc tinh
df = df.rename(columns = {"Engine HP": "HP"})
df.head(5)
```

```python
#kiem tra dong, cot
df.shape
```

```python
#Kiem tra so luong dong du lieu bi trung
duplicate_rows_df = df[df.duplicated()]
print("So luong dong bi trung: ", duplicate_rows_df.shape)
```

```python
#Dem so dong du lieu
df.count()
```

```python
#Xoa cac dong du lieu bi trung
df = df.drop_duplicates()
df.head(5)
```

```python
df.count()
```

```python
#Kiem tra gia tri null
print(df.isnull().sum())
```

```python
#Xoa cac gia tri null
df = df.dropna()
df.count()
```

```python
print(df.isnull().sum())
```

```python
#Kiem tra gia tri outliners cua thuoc tinh Cylinders
sns.boxplot(x = df['Engine Cylinders'])
```

```python
#Ve bieu do histogram
df.Make.value_counts().nlargest(40).plot(kind = 'bar', figsize = (10,5))
plt.title("So luong xe theo hang san xuat")
plt.ylabel('So luong xe')
plt.xlabel('Hang xe')
```

```python
plt.show()
```

```python
#ve bieu do heatmaps
plt.figure(figsize = (10,5))
c = df.select_dtypes(include = ['number']).corr()
sns.heatmap(c,cmap = "BrBG", annot = True)
plt.show()
```

```python
#Ve bieu do scatterplot giua thuoc tinh HP va Price
fig, ax = plt.subplots(figsize = 10,6)
ax.scatter(df['HP'], df['Price'])
ax.set_xlabel('HP')
ax.set_ylabel('Price')
plt.show()
```
