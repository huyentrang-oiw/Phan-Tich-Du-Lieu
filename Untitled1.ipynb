```python
#Ve bieu do line 
#Vi du bieu do tieu thu dien nang

import matplotlib.pyplot as plt
year = [1972, 1982, 1992, 2002, 2012]
VietNam = [120, 178, 205, 395, 724]
USA = [10, 25, 58, 119, 274]

plt.plot (year, VietNam, color = 'red', label = 'Viet Nam')

plt.plot (year, USA, color = 'blue', label = 'Hoa Ky')

plt.xlabel ('Nam')
plt.ylabel('Luong dien tieu thu')

plt.title('So sanh luong dien tieu thu dien')

plt.legend()
plt.show()
```

```python
#Bieu do co nhieu cot de so sanh
import numpy as np
import matplotlib.pyplot as plt
X = ['Nhm A','Nhm B','Nhm C','Nhm D']
nu = [10,20,20,40]
nam = [20,30,25,30]
X_axis = np.arange(len(X)) # Ly cc gi tr  hin th cc nhm
plt.bar(X_axis - 0.2, nu, width=0.4, label = 'N')
plt.bar(X_axis + 0.2, nam, width=0.4, label = 'Nam')
plt.xticks(X_axis, X) # X_axis l v tr, X l tn s hin th
plt.xlabel("Nhm")
plt.ylabel("S lng sinh vin")
plt.title("S sinh vin ca tng nhm")
plt.legend()
plt.show()
```

```python
#Bieu do Histogram
import matplotlib.pyplot as plt
import numpy as np
data = np.random.randn(1000)
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.xlabel('S lng')
plt.ylabel('Tn sut')
plt.title('Biu  Histogram')
plt.show()
```

```python
#Bieu do histogram voi Matplotlib va Seaborn
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
data = np.random.randn(1000)
sns.histplot(data, bins=30, kde=True, color='y', edgecolor='red')
plt.xlabel('S lng')
plt.ylabel('Tn sut')
plt.title('Biu  Histogram')
plt.show()
```

```python
#Scatterplot Bieu do scatter voi 2 class
import matplotlib.pyplot as plt
import numpy as np
x = np.array([5,7,8,7,2,17,2,9])
y = np.array([99,86,87,88,111,86,103,87])
plt.scatter(x, y, s = 50)
x = np.array([2,2,8,1,15,8,12,9,7,11])
y = np.array([100,105,84,105,90,99,90,91,80,85])
plt.scatter(x, y)
plt.show()
```

```python
#Bieu do scatter voi tham so colormap
import matplotlib.pyplot as plt
import numpy as np
x = np.array([5,7,8,7,2,17,2,9])
y = np.array([99,86,87,88,111,86,103,87])
colors = np.array([20, 30, 40, 45, 50, 55, 60, 70])
plt.scatter(x, y, c=colors, cmap='viridis')
plt.colorbar()
plt.show()
```

```python
#scatterplot voi tham so nang cao
import matplotlib.pyplot as plt
import numpy as np
x = np.array([9,5,7,8,7,2,17,2,9,12,32,45,21,45,10])
y = np.array([99,86,87,88,111,86,103,87,33,45,76,87,65,80,77])
colors = np.random.rand(15)
sizes = 100 * np.random.rand(15)
plt.scatter(x, y, c=colors, s=sizes, alpha=0.7, cmap='viridis')
plt.title("Tiu  biu  Scatterplot")
plt.xlabel("Trc X")
plt.ylabel("Trc Y")
plt.colorbar(label='Bng mu')
plt.show()
```

```python
#Bieu do pie
import matplotlib.pyplot as plt
monhoc = ['CSDL', 'MMT', 'QTM','AI', 'KTLT', 'CSTL']
data = [23, 17, 35, 29, 12, 41]
plt.pie(data, labels = monhoc)
plt.show()
```

```python
#Bieu do pie voi tham so nang cao
import matplotlib.pyplot as plt
import numpy as np
data = [23, 17, 35, 29, 12, 41]
monhoc = ['CSDL', 'MMT', 'QTM','AI', 'KTLT', 'CSTL']
mausac = ["black", "hotpink", "b", "#4CAF50","yellow","red"]
vitri = [0.2, 0, 0, 0, 0, 0]
plt.pie(data, labels = monhoc, colors = mausac, explode = vitri, shadow = True)
plt.legend()
plt.show()
```

```python
#Bieu do boxplot
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(10)
data = np.random.normal(100, 20, 200)
plt.boxplot(data) # Tham s vert=False s v boxplot nm ngang
plt.show()
```

```python
#Ve nhieu boxplot
import matplotlib.pyplot as plt
import numpy as np
data_1 = np.random.normal(100, 10, 200)
data_2 = np.random.normal(90, 20, 200)
data_3 = np.random.normal(80, 30, 200)
data_4 = np.random.normal(70, 40, 200)
data = [data_1, data_2, data_3, data_4]
fig = plt.figure(figsize =(5, 4)) #To 1 figure vi kch thc 5x4
ax = fig.add_axes([0, 0, 1, 1])
ax.boxplot(data, vert=False)
plt.show()
```

```python
#De ve nhieu bieu do can su dung ham subplot() C php: subplot(nrow,ncol, index)
x = np.arange(1,9)
y = x*3
plt.subplot(2,1,1)
plt.plot([1,2,3,4],[2,4,9,16])
plt.title("Hnh 1")
plt.subplot(2,1,2)
plt.plot(x,y,"r^")
plt.title("Hnh 2")
plt.show()
```

```python
#Ve nhieu bieu do tren cung mot fig
import matplotlib.pyplot as plt
fig, bieudo = plt.subplots(2, 2)
bieudo[0, 0].plot([1, 2, 3], [4, 5, 6]) #V ti v tr dng 0, ct 0
bieudo[0, 1].scatter([1, 2, 3], [4, 5, 6]) #V ti v tr dng 0, ct 1
bieudo[1, 0].bar([1, 2, 3], [4, 5, 6]) #V ti v tr dng 1, ct 0
bieudo[1, 1].hist([1, 2, 2, 3, 3, 3, 3, 3, 3, 4, 4, 5]) #V ti v tr dng 1, ct 1
plt.show()
```
